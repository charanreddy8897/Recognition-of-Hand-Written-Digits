```python
#Add each of rows of the digits to their respective lists
f = open('Dataset.csv')

one = []
four = []
two = []
six = []
zero = []

count = 0
for i in f:
    s = i.strip()
    
    if s[len(s)-1] == '1':
        one.append(s)
    if s[len(s)-1] == '4':
        four.append(s)
    if s[len(s)-1] == '2':
        two.append(s)
    if s[len(s)-1] == '6':
        six.append(s)
    if s[len(s)-1] == '0':
        zero.append(s)
    
    count += 1
f.close()



#Take the minimum length of all 5 digit lists and create a list of lists containing the testing sets
length = min(len(one), len(four), len(two), len(six), len(zero))

out = []

i = 0
while i < length:
    out.append([one[i],four[i],two[i],six[i],zero[i]])
    i += 1
	
	
	
#Output all the testing sets to their own respective file in a folder
import os
import sys

save_path = 'output'
file_name = ""

c = 1
for j in out:
    file_name = str(c) + ".csv"
    completeName = os.path.join(save_path, file_name)
    file1 = open(completeName, "w")
    for k in j: 
        file1.write(k + "\n")
    file1.close()
    c += 1



#Print the count for each digit
print(f'Total One Digits: {len(one)}')
print(f'Total Four Digits: {len(four)}')
print(f'Total Two Digits: {len(two)}')
print(f'Total Six Digits: {len(six)}')
print(f'Total Zero Digits: {len(zero)}')



#Output all the training sets to their own respective file in a folder
c = 1
save_path = 'output2'
file_name = ""

for j in out:
    file_name = str(c) + ".csv"
    completeName = os.path.join(save_path, file_name)
    file1 = open(completeName, "w")
    file2 = open('Dataset.csv')
    for k in file2: 
        if k.strip() not in j:
            file1.write(k.strip() + '\n')
    file2.close()
    file1.close()
    c += 1
```
